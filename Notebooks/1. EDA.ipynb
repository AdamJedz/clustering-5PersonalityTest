# Answers by category

## Most and least agreeable questions

```python
df_answers = df.filter(regex='(\w{3}\d+)$')
```

```python
mean_answer = df_answers.mean().sort_values(ascending=False)
mean_answer[:5]
```

Most agreeable questions were:  
* **OPN9** - I spend time reflecting on things.  
* **OPN3** - I have a vivid imagination.  
* **OPN7** - I am quick to understand things.  
* **CSN3** - I pay attention to details.  
* **OPN10** - I am full of ideas.

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=mean_answer[:10].index, y=mean_answer[:10].values)
plt.title('10 most agreeable questions')
plt.xlabel('Questions')
plt.ylabel('Average answer')
plt.ylim(top=5)
```

```python
mean_answer[-5:].sort_values()
```

Least agreeable questions were:  
* **OPN6** - I do not have a good imagination.  
* **OPN4** - I am not interested in abstract ideas.  
* **OPN2** - I have difficulty understanding abstract ideas.  
* **AGR7** - I am not really interested in others.  
* **AGR1** - I feel little concern for others.

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=mean_answer[-10:].sort_values().index, y=mean_answer[-10:].sort_values().values)
plt.title('10 least agreeable questions')
plt.xlabel('Questions')
plt.ylabel('Average answer')
plt.ylim(top=5)
```

## Agreeability by question category

```python
import re
categories = list(set([re.findall('^\w{3}', col)[0] for col in df_answers.columns]))
```

```python
out = dict()
for cat in categories:
    tmp = df_answers.copy()
    tmp = tmp.filter(regex='^' + cat)
    out[cat] = tmp.mean(axis=1).mean()
out
```

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=list(out.keys()), y = list(out.values()))
plt.title('Agreeability by question category')
plt.xlabel('Question categories')
plt.ylabel('Average answer')
plt.ylim(top=5)
```

# Average response time per category 

```python
df_time = df.filter(regex='(\w{3}\d+)_E$')
```

```python
mean_time = df_time.mean().sort_values(ascending=False)
mean_time[:6]
```

Questions with longest response time were:

* **CSN1** - I am always prepared.  
* **OPN2** - I have difficulty understanding abstract ideas.  
* **CSN8** - I shirk my duties.

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=mean_time[:10].index, y=mean_time[:10].values)
plt.title('10 questions with longest response time')
plt.xlabel('Questions')
plt.ylabel('Response time')
```

```python
mean_time[-5:].sort_values()
```

Questions with shortest response time:  
* **EST10** - I often feel blue.  
* **OPN10** - I am full of ideas.  
* **EST9** - I get irritated easily.  
* **OPN8** - I use difficult words.  
* **CSN9** - I follow a schedule.

```python
out = dict()
for cat in categories:
    tmp = df_time.copy()
    tmp = tmp.filter(regex='^' + cat)
    out[cat] = tmp.mean(axis=1).mean()
out
```

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=list(out.keys()), y = list(out.values()))
plt.title('Average response time by question category')
plt.xlabel('Question categories')
plt.ylabel('Average response time')
```
